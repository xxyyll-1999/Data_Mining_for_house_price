```python
#利用fillna的方法填充的缺失值
for column in df_all.select_dtypes(include=['object']).columns.tolist():
    df_all[column] = df_all[column].fillna('Missing')

df_all.select_dtypes(include=['object']).isnull().sum()
```

```python
#使用交叉验证来更好地进行评估
from sklearn.model_selection import cross_val_score

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
#使用梯度提升建模
from sklearn.ensemble import GradientBoostingRegressor
gbrt=GradientBoostingRegressor(random_state=42)
gbrt.fit(X_train,y_train)
```

```python
gbrt_scores=cross_val_score(gbrt,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
gbrt_rmse_scores=np.sqrt(-gbrt_scores)
display_scores(gbrt_rmse_scores)
```

```python
#梯度提升看起来有戏
```

```python
#使用随机森林建模
from sklearn.ensemble import RandomForestRegressor
rnd_reg=RandomForestRegressor(random_state=42)
rnd_reg.fit(X_train,y_train)
rnd_forest_scores=cross_val_score(rnd_reg,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
rnd_forest_rmse_scores=np.sqrt(-rnd_forest_scores)
display_scores(rnd_forest_rmse_scores)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import warnings #因为Lasso回归会报错，为了不影响结果的美观
warnings.filterwarnings('ignore')

pipelines =  {
            'gbrt':make_pipeline(StandardScaler(),gbrt),
            'rnd':make_pipeline(StandardScaler(), rnd_reg),
            'extra_trees':make_pipeline(StandardScaler(), extra_trees_reg),
            'lasso': make_pipeline(StandardScaler(), lasso_reg),
            'ridge': make_pipeline(StandardScaler(), ridge_reg),
                }  
```

```python
for reg_name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)
    pred = pipeline.predict(X_test)
    mse=mean_squared_error(y_test, pred)
    r2=r2_score(y_test,pred)
    print(reg_name,np.sqrt(mse),r2)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=1500, random_state=42)
gbrt.fit(X_train,y_train)

errors = [mean_squared_error(y_test, y_pred)
          for y_pred in gbrt.staged_predict(X_test)]
bst_n_estimators = np.argmin(errors) + 1

gbrt_best = GradientBoostingRegressor(max_depth=2,n_estimators=bst_n_estimators, random_state=42)
gbrt_best.fit(X_train, y_train)
```
