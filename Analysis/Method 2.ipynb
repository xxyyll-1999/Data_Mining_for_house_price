```python
#筛选有效模型
```

```python
#使用交叉验证来更好地进行评估
from sklearn.model_selection import cross_val_score

def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
#使用梯度提升建模
from sklearn.ensemble import GradientBoostingRegressor
gbrt=GradientBoostingRegressor(random_state=42)
gbrt.fit(X_train,y_train)
```

```python
gbrt_scores=cross_val_score(gbrt,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
gbrt_rmse_scores=np.sqrt(-gbrt_scores)
display_scores(gbrt_rmse_scores)
```

```python
#梯度提升看起来有戏
```

```python
#使用随机森林建模
from sklearn.ensemble import RandomForestRegressor
rnd_reg=RandomForestRegressor(random_state=42)
rnd_reg.fit(X_train,y_train)
rnd_forest_scores=cross_val_score(rnd_reg,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
rnd_forest_rmse_scores=np.sqrt(-rnd_forest_scores)
display_scores(rnd_forest_rmse_scores)
```

```python
#使用极端随机树建模
from sklearn.ensemble import ExtraTreesRegressor
extra_trees_reg = ExtraTreesRegressor(random_state=42)
extra_trees_reg.fit(X_train,y_train)
extra_trees_scores=cross_val_score(extra_trees_reg,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
extra_trees_rmse_scores=np.sqrt(-extra_trees_scores)
display_scores(extra_trees_rmse_scores)
```

```python
#采用线性回归建模
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
lin_scores=cross_val_score(lin_reg,X_train,y_train,
                                scoring="neg_mean_squared_error", cv=5)
lin_rmse_scores=np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```

```python
#线性回归模型看起来就不太行
```

```python
#使用索套回归建模
import warnings #因为Lasso回归会报错，为了不影响结果的美观
warnings.filterwarnings('ignore')
from sklearn.linear_model import Lasso
lasso_reg=Lasso(random_state=42)
lasso_reg.fit(X_train,y_train)
lasso_scores=cross_val_score(lasso_reg,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
lasso_rmse_scores=np.sqrt(-lasso_scores)
display_scores(lasso_rmse_scores)
```

```python
#lasso回归也看起来有戏
```

```python
#使用岭回归建模
from sklearn.linear_model import Ridge
ridge_reg=Ridge(random_state=42)
ridge_reg.fit(X_train,y_train)
ridge_scores=cross_val_score(ridge_reg,X_train,y_train,scoring="neg_mean_squared_error", cv=5)
ridge_rmse_scores=np.sqrt(-ridge_scores)
display_scores(ridge_rmse_scores)
```

```python
#选择有希望的模型进行优化
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import warnings #因为Lasso回归会报警告，为了不影响结果的美观
warnings.filterwarnings('ignore')

pipelines =  {
            'gbrt':make_pipeline(StandardScaler(),gbrt),
            'rnd':make_pipeline(StandardScaler(), rnd_reg),
            'extra_trees':make_pipeline(StandardScaler(), extra_trees_reg),
            'lasso': make_pipeline(StandardScaler(), lasso_reg),
            'ridge': make_pipeline(StandardScaler(), ridge_reg),
                }  
```

```python
for reg_name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)
    pred = pipeline.predict(X_test)
    mse=mean_squared_error(y_test, pred)
    r2=r2_score(y_test,pred)
    print(reg_name,np.sqrt(mse),r2)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(max_depth=2, n_estimators=1500, random_state=42)
gbrt.fit(X_train,y_train)

errors = [mean_squared_error(y_test, y_pred)
          for y_pred in gbrt.staged_predict(X_test)]
bst_n_estimators = np.argmin(errors) + 1

gbrt_best = GradientBoostingRegressor(max_depth=2,n_estimators=bst_n_estimators, random_state=42)
gbrt_best.fit(X_train, y_train)
```

```python
bst_n_estimators
```
